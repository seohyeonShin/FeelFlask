```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open('../flavor_update.json', 'r') as fr:
    flavor_info = json.load(fr)
```

```python
flavor_info
```

```python
# flavor Ïù¥Î¶ÑÏùÑ keyÎ°ú, Í∞Å flavor Îßõ Ï†ïÎ≥¥Î•º valueÎ°ú ÌïòÎäî dictionary ÏÉùÏÑ±
ind_to_flavor = {}
index = 1
for flavor in flavor_info:
    element = {}
    for key in flavor:
        if key != 'name':
            element[key] = flavor[key]
    ind_to_flavor[index] = element
    index+=1
```

```python
ind_to_flavor
```

```python
flavor_dic = {}
for flavor in flavor_info:
    element = {}
    for feature, value in flavor.items():
        if feature != "name":
            element[feature] = value
    flavor_dic[flavor['name']] = element
```

```python
flavor_dic
```

```python
user_profile = {'ABV': 30,
                'sweet' : 75,
                'sour' : 60,
                'bitter' : 45,
                'spicy' : 30,
                'herbal' : 50,
                'floral' : 45,
                'fruity' : 25,
                'nutty' : 50,
                'boozy' : 35,
                'astringent' : 50,
                'umami' : 35,
                'salty' : 35,
                'Perceived_temperature' : 25,
                'creamy' : 25,
                'smoky' : 25,
                }

sorted_values = []
for item in user_profile.items():
    sorted_values.append(item)
sorted_values = sorted(sorted_values, key=lambda x: x[1], reverse=True)
# Í∞ÄÏû• ÌäπÏßïÏ†ÅÏù∏ 3Í∞úÏùò featureÍ∞íÏùÑ Í∞ÄÏßÄÎäî featureÎì§ÏùÑ Ï∂îÏ∂úÌï¥ ÎÉÖÎãàÎã§.
top_3_features = sorted_values[:3]
```

```python
top_3_features
```

```python
# lower is better
filter_score = {}
for ing_name, feature_list in flavor_dic.items():
    filter_score[ing_name] = 0
    for feature, value in top_3_features:
        filter_score[ing_name] += abs(feature_list[feature] - value)
```

```python
filter_score
```

```python
sorted_ingredient = []
for item in filter_score.items():
    sorted_ingredient.append(item)
sorted_ingredient = sorted(sorted_ingredient, key=lambda x: x[1])
top_10_ingredient = []

# Ï†ÑÏ≤¥ Ïû¨Î£åÍ∞Ä ÏïÑÎãå 10Í∞ú ÍπåÏßÄÏùò Ïû¨Î£åÎ•º Ï∂îÎ¶ΩÎãàÎã§.
for ing_name, score in sorted_ingredient[:10]:
    top_10_ingredient.append(ing_name)
top_10_flavor = {}
for ing_name in top_10_ingredient:
    top_10_flavor[ing_name] = {}
    for feature, value in flavor_dic[ing_name].items():
        top_10_flavor[ing_name][feature] = value
```

```python
flavor_info
boozy = {}
sweet = {}
sour = {}
bitter = {}
umami = {}
salty = {}
astringent = {}
perceived_t = {}
spicy = {}
herbal = {}
floral = {}
fruity = {}
nutty = {}
creamy = {}
smoky = {}
count = 0
for cocktail in flavor_info:
    boozy[count] = cocktail['boozy']
    sweet[count] = cocktail['sweet']
    sour[count] = cocktail['sour']
    bitter[count] = cocktail['bitter']
    umami[count] = cocktail['umami']
    salty[count] = cocktail['salty']
    astringent[count] = cocktail['astringent']
    perceived_t[count] = cocktail['Perceived_temperature']
    spicy[count] = cocktail['spicy']
    herbal[count] = cocktail['herbal']
    floral[count] = cocktail['floral']
    fruity[count] = cocktail['fruity']
    nutty[count] = cocktail['nutty']
    creamy[count] = cocktail['creamy']
    smoky[count] = cocktail['smoky']
    count += 1
```

```python
sorted_boozy = sorted(boozy.items(), key=lambda x: x[1], reverse=True)
sorted_sweet = sorted(sweet.items(), key=lambda x: x[1], reverse=True)
sorted_sour = sorted(sour.items(), key=lambda x: x[1], reverse=True)
sorted_bitter = sorted(bitter.items(), key=lambda x: x[1], reverse=True)
sorted_umami = sorted(umami.items(), key=lambda x: x[1], reverse=True)
sorted_salty = sorted(salty.items(), key=lambda x: x[1], reverse=True)
sorted_astringent = sorted(astringent.items(), key=lambda x: x[1], reverse=True)
sorted_perceived_t = sorted(perceived_t.items(), key=lambda x: x[1], reverse=True)
sorted_spicy = sorted(spicy.items(), key=lambda x: x[1], reverse=True)
sorted_herbal = sorted(herbal.items(), key=lambda x: x[1], reverse=True)
sorted_floral = sorted(floral.items(), key=lambda x: x[1], reverse=True)
sorted_fruity = sorted(fruity.items(), key=lambda x: x[1], reverse=True)
sorted_nutty = sorted(nutty.items(), key=lambda x: x[1], reverse=True)
sorted_creamy = sorted(creamy.items(), key=lambda x: x[1], reverse=True)
sorted_smoky = sorted(smoky.items(), key=lambda x: x[1], reverse=True)
```

```python
top_n = 30
top_n_list = []
below_n_list = []
sorted_feature = [sorted_boozy, sorted_sweet, sorted_sour, sorted_bitter,
                  sorted_umami, sorted_salty, sorted_astringent, 
                  sorted_perceived_t, sorted_spicy, sorted_herbal, sorted_floral, 
                  sorted_fruity, sorted_nutty, sorted_creamy, sorted_smoky]
feature_name = ['boozy', 'sweet', 'sour', 'bitter', 'umami', 'salty',
                'astringent', 'perceived_t', 'spicy', 'herbal', 'floral',
                'fruity', 'nutty', 'creamy', 'smoky']
ind = 0
for feature in sorted_feature:
    top_n_list.append(f'//// - {feature_name[ind]} - ////')
    below_n_list.append(f'//// - {feature_name[ind]} - ////')
    for i in range(top_n):
        top_n_list.append(name_mapper[feature[i][0]])
        below_n_list.append(name_mapper[feature[(i+1)*-1][0]])
    ind += 1
```
